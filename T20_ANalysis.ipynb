#### 2. In-Depth Analysis: Examine player statistics, including runs, wickets, and over-by-over progress, to understand the dynamics of the game.

#### Common statistics

```python
# batsman1_statistics
batsman1_run = dataset.groupby('batsman1_name')['runs'].sum().reset_index()
batsman1_run = pd.DataFrame(batsman1_run)
batsman1_run.describe()
```

```python
#batsman 2 statistics
batsman2_run = dataset.groupby('batsman2_name')['runs'].sum().reset_index()
batsman2_run = pd.DataFrame(batsman2_run)
batsman2_run.describe()
```

```python
#bowler 1 statistics
bowler1_wickets = dataset.groupby('bowler1_name')['bowler1_wkts'].sum().reset_index()
bowler1_wickets = pd.DataFrame(bowler1_wickets)
bowler1_wickets.describe()
```

```python
#bowler 2 statistics
bowler2_wickets = dataset.groupby('bowler2_name')['bowler2_wkts'].sum().reset_index()
bowler2wickets = pd.DataFrame(bowler2_wickets)
bowler2wickets.describe()
```

#### Team performance statistics

```python
current_innings = dataset.groupby('current_innings')['runs'].sum().reset_index()
current_innings = pd.DataFrame(current_innings)
current_innings.describe()
```

```python
batsman_runs = dataset.groupby('batsman1_name')['batsman1_runs'].sum() + dataset.groupby('batsman2_name')['batsman2_runs'].sum()
batsman_balls_faced = dataset.groupby('batsman1_name')['batsman1_balls'].sum() + dataset.groupby('batsman2_name')['batsman2_balls'].sum()

bowler_overs_bowled = dataset.groupby('bowler1_name')['bowler1_overs'].sum() + dataset.groupby('bowler2_name')['bowler2_overs'].sum()
bowler_maidens_bowled = dataset.groupby('bowler1_name')['bowler1_maidens'].sum() + dataset.groupby('bowler2_name')['bowler2_maidens'].sum()
bowler_runs_conceded = dataset.groupby('bowler1_name')['bowler1_runs'].sum() + dataset.groupby('bowler2_name')['bowler2_runs'].sum()
bowler_wickets_taken = dataset.groupby('bowler1_name')['bowler1_wkts'].sum() + dataset.groupby('bowler2_name')['bowler2_wkts'].sum()

batsman_stats = pd.DataFrame({'Runs': batsman_runs, 'Balls Faced': batsman_balls_faced})
batsman_stats = batsman_stats.sort_values(by='Runs', ascending=False).head(10)
print("Batsman Statistics:")
print(batsman_stats)

bowler_stats = pd.DataFrame({'Overs Bowled': bowler_overs_bowled, 'Maidens': bowler_maidens_bowled,
                             'Runs Conceded': bowler_runs_conceded, 'Wickets Taken': bowler_wickets_taken})
bowler_stats = bowler_stats.sort_values(by='Overs Bowled', ascending=False).head(10)
print("\nBowler Statistics:")
print(bowler_stats)
```

#### 3. Event Inference: Identify and analyze critical match events, such as boundaries, wickets, and strategic moments. Uncover the factors that contribute to the success or downfall of a team.

```python
boundary_events = dataset[(dataset['runs'] == 4) | (dataset['runs'] == 6)]
wide_no_ball_events = dataset[(dataset['isWide'] == True) | (dataset['isNoball'] == True)]
wkt_events = dataset[dataset['wkt_bowler_name'].notnull()]
print("Number of wicket events:", len(wkt_events))
print("Number of boundary events:", len(boundary_events))
print("Number of wide/no-ball events:", len(wide_no_ball_events))
```

#### 4. Performance Evaluation: Evaluate individual and team performances, including standout players, consistent run-scorers, and top wicket-takers. Discover what sets them apart.

```python
batsman_runs = dataset.groupby('batsman1_name')['batsman1_runs'].sum() + dataset.groupby('batsman2_name')['batsman2_runs'].sum()
bowler_wickets = dataset.groupby('bowler1_name')['bowler1_wkts'].sum() + dataset.groupby('bowler2_name')['bowler2_wkts'].sum()

top_batsmen = batsman_runs.nlargest(10)
top_bowlers = bowler_wickets.nlargest(10)

print("Top 10 Batsmen by Runs:")
print(top_batsmen)

print("\nTop 10 Bowlers by Wickets:")
print(top_bowlers)
```

#### 5. Statistical Insights: Utilize statistical techniques to gain insights into team strategies, batting, bowling, and match outcomes. Create visualizations that convey the story of the tournament.

```python
#batiing performance of batsman1_runs

top_10_runs = batsman1_run.sort_values(by = 'runs', ascending=False ).head(10)
plt.figure(figsize=(14,7))
sns.barplot(x='batsman1_name', y = 'runs', data = top_10_runs)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#batsman 2

top_10_runs_batsman2 = batsman2_run.sort_values(by = 'runs', ascending=False ).head(10)
plt.figure(figsize=(14,7))
sns.barplot(x='batsman2_name', y = 'runs', data = top_10_runs_batsman2)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python

top_10_runs_bowler1_wickets = bowler1_wickets.sort_values(by = 'bowler1_wkts', ascending=False ).head(10)
plt.figure(figsize=(14,7))
sns.barplot(x='bowler1_name', y = 'bowler1_wkts', data = top_10_runs_bowler1_wickets)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#bowler 2

top_10_runs_bowler2_ickets = bowler2wickets.sort_values(by = 'bowler2_wkts', ascending=False ).head(10)
plt.figure(figsize=(14,7))
sns.barplot(x='bowler2_name',y = 'bowler2_wkts', data = top_10_runs_bowler2_ickets)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#team Performance - batting

top_10_runs_current_innings = current_innings.sort_values(by = 'runs', ascending=False ).head(10)
plt.figure(figsize=(14,7))
sns.barplot(x='current_innings',y = 'runs', data = top_10_runs_current_innings)
plt.xticks(rotation=45, ha='right')
plt.show()
```
